**Load Necesary libraries** 

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

### BasicGraph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize your graph
plt.figure(figsize=(5, 3), dpi=150)
# plt.plot(x, y, label='Line 1', color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)
# plt.plot(x, y, label='2x', color='red', linewidth=2, marker='o', markersize=8, linestyle='--', markeredgecolor='black', markerfacecolor='yellow')

# Shortened notatioan
plt.plot(x, y, 'b^--', label='2x', linewidth=2, markersize=8, markeredgecolor='black', markerfacecolor='yellow')

# Line Number 2
x2 = np.arange(0, 4.5, 0.5)
# print(x2)

# plot part of graph as line
plt.plot(x2[:6], x2[:6]**2, label='x^2', color='r', marker='s', markersize=6, linewidth=1.5)

# plot remainder of graph as dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# X and Y axis tickmarks
plt.xticks([0,1,2, 3, 4])
plt.yticks([0, 2, 4, 6, 8,10])
# plt.grid(True, linestyle='--', color='gray', alpha=0.5)

# Add Title
plt.title("Our First Graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'blue'})

# Add Labels
plt.xlabel("X-axis", fontdict={'fontname': 'Arial', 'fontsize': 14, 'color': 'green'})
plt.ylabel("Y-axis", fontdict={'fontname': 'Arial', 'fontsize': 14, 'color': 'green'})

# Adding a legend
plt.legend()
plt.show()
```

### BarChart

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

bars = plt.bar(labels, values, color=['red', 'blue', 'green', 'orange'], edgecolor='black')


bars[0].set_hatch('x')  # Hatching for the first bar
bars[1].set_hatch('*')  # Hatching for the second bar
bars[2].set_hatch('o')  # Hatching for the third bar
bars[3].set_hatch('/')  # Hatching for the fourth bar

plt.title("Bar Chart Example", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'blue'})
plt.xlabel("Categories", fontdict={'fontname': 'Arial', 'fontsize': 14, 'color': 'green'})
plt.ylabel("Values", fontdict={'fontname': 'Arial', 'fontsize': 14  , 'color': 'green'})
plt.xticks(rotation=45)         

plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()      
```

### Pie Chart

```python
# Example of a pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']        
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']  
explode = (0.1, 0, 0, 0)  # explode the first slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie Chart Example", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'blue'})
plt.show()
```
