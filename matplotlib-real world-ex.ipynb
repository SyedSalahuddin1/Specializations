#### Load Necessary libraries

### Real World Examples

```python
gas = pd.read_csv('./data/gas_prices.csv')

gas
```

```python
plt.figure(figsize=(8, 5))
plt.title("Gas Prices Overtime")

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# plt.plot(gas.Year, gas.Mexico, 'g.-', label='Mexico')
plt.plot(gas.Year, gas['South Korea'], 'y.-' ,label='South Korea')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Gas Prices Overtime", fontdict={'fontweight':'bold', 'fontsize':20})

# plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# # plt.plot(gas.Year, gas.Mexico, 'g.-', label='Mexico')
# plt.plot(gas.Year, gas['South Korea'], 'y.-' ,label='South Korea')

countries_to_look_at = ['Mexico', 'USA', 'Canada', 'South Korea', 'Italy' ]

for country in gas:
    if country  in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker = '.', label=country)

# print(gas.Year[::3])

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('./data/fifa_data.csv')

fifa.head()
```

#### Histograms

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa['Overall'], bins=bins, color='#B32407', edgecolor='black', alpha=0.7)
plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of FIFA Player Skill Levels", fontdict={'fontweight':'bold', 'fontsize':20})


plt.show()
```

#### Pie Charts

```python
from turtle import right


left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels=['Left', 'Right'], autopct='%1.1f%%', startangle=90, colors=['#B32407', '#F2A900'], explode=(0.1, 0))
plt.title("Foot Percentage of FIFA Players")

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = (fifa.Weight < 125).sum()
light_medium = ((fifa.Weight >= 125) & (fifa.Weight < 150)).sum()
medium = ((fifa.Weight >= 150) & (fifa.Weight < 175)).sum()
medium_heavy = ((fifa.Weight >= 175) & (fifa.Weight < 200)).sum()
heavy = (fifa.Weight >= 200).sum()


weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')


plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.show()
```
