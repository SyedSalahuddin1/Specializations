#### Load Necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Real World Examples

```python
gas = pd.read_csv('./data/gas_prices.csv')

gas
```

```python
plt.figure(figsize=(8, 5))
plt.title("Gas Prices Overtime")

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# plt.plot(gas.Year, gas.Mexico, 'g.-', label='Mexico')
plt.plot(gas.Year, gas['South Korea'], 'y.-' ,label='South Korea')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Gas Prices Overtime", fontdict={'fontweight':'bold', 'fontsize':20})

# plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# # plt.plot(gas.Year, gas.Mexico, 'g.-', label='Mexico')
# plt.plot(gas.Year, gas['South Korea'], 'y.-' ,label='South Korea')

countries_to_look_at = ['Mexico', 'USA', 'Canada', 'South Korea', 'Italy' ]

for country in gas:
    if country  in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker = '.', label=country)

# print(gas.Year[::3])

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```
