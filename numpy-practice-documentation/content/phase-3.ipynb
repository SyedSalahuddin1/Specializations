## Advance Operation with business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape: ", sales_data.shape)
print("\n Sample data for 1st 3 restaurant: ",sales_data[:, 1:])
```

```python
# total sales per year
print("SUM: ",np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("Yearly Total: ",yearly_total)

minimum_sales = np.min(sales_data[:,1:], axis=1)
print("Minimum Sales: ",minimum_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Max Sales: ",max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales: ",avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across  all restaurents")
plt.ylabel("Sales")
plt.xlabel("Years")
plt.grid(True)
plt.show()
```

```python
from numpy import angle, vectorize


vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("Vecor Multiplication: ", vector1*vector2)

print("Dot product", np.dot(vector1, vector2))

angle = np.arccos(
    np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
)
print("Angle in radians: ",angle)

restaurent_types = np.array(['birayni','chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized upper", vectorized_upper(restaurent_types))
```
