```python
bios["hieght_category"] = bios["height_cm"].apply(lambda x: 'Short' if x < 165 else ("Average" if x < 185 else "Tall"))
```

```python
bios.head()
```

```python
def categorize_athlete(row):
    if row["height_cm"] < 175 and row["weight_kg"] < 70:
        return "Lightweight"
    elif row["height_cm"] < 185 and row["weight_kg"] <= 80:
        return "Middleweight"
    else:
        return "Heavyweight"
    
bios["Category"] = bios.apply(categorize_athlete, axis=1)
```
